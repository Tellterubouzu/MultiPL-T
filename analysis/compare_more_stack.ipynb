```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bline = {
    "ocaml" : 0.015, 
    "racket" : 0.047,
    "lua" : 0.121,
}
# Read data
ocaml_stack = list(pd.read_csv('../experiments/ocaml_stack_full_1b/results.csv')["Estimate"])
rkt_stack = list(pd.read_csv('../experiments/rkt_stack_full_1b/results.csv')["Estimate"])
lua_stack = list(pd.read_csv('../experiments/lua_stack_full_1b/results.csv')["Estimate"])

lua_multiplt = list(pd.read_csv('../experiments/lua_full_1b/results.csv')["Estimate"])
rkt_multiplt = list(pd.read_csv('../experiments/rkt_full_1b/results.csv')["Estimate"])
ocaml_multiplt = list(pd.read_csv('../experiments/ocaml_full_1b/results.csv')["Estimate"])
```

```python
# Plot 1
plt.figure(figsize=(10, 5))
xpos = list(range(1,8))
plt.plot(xpos, ocaml_stack, label="OCaml", color="red")
plt.plot(xpos, ocaml_multiplt, label="OCaml (multiplt)", color="red", linestyle="--")
xpos = list(range(1,7))
plt.plot(xpos, rkt_stack, label="Racket", color="blue")
plt.plot(xpos, lua_stack, label="Lua", color="green")
plt.plot(xpos, rkt_multiplt, label="Racket (multiplt)", color="blue", linestyle="--")
plt.plot(xpos, lua_multiplt, label="Lua (multiplt)", color="green", linestyle="--")


# set x-axis 1-7
plt.xlabel("Epoch Number")
plt.ylabel("HumanEval pass@1")
handles, labels = plt.gca().get_legend_handles_labels()
order = [0,1, 2, 4, 3, 5]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
        fontsize=9, loc="upper right")
plt.show()
```

```python
#baseline
plt.figure(figsize=(10, 5))
plt.plot([0], [bline["ocaml"]], label="OCaml (baseline)", color="red", marker="o")
plt.plot([0], [bline["racket"]], label="Racket (baseline)", color="blue", marker="o")
plt.plot([0], [bline["lua"]], label="Lua (baseline)", color="green", marker="o")

# Plot 2
xpos = list(range(1,8))
plt.plot(xpos, ocaml_stack, label="OCaml", color="red")
plt.plot(xpos, ocaml_multiplt, label="OCaml (multiplt)", color="red", linestyle="--")
xpos = list(range(1,7))
plt.plot(xpos, rkt_stack, label="Racket", color="blue")
plt.plot(xpos, lua_stack, label="Lua", color="green")
plt.plot(xpos, rkt_multiplt, label="Racket (multiplt)", color="blue", linestyle="--")
plt.plot(xpos, lua_multiplt, label="Lua (multiplt)", color="green", linestyle="--")


# set x-axis 1-7
plt.xlabel("Epoch Number")
plt.ylabel("HumanEval pass@1")
handles, labels = plt.gca().get_legend_handles_labels()
order = [0, 3, 4, 1, 5, 7, 2, 6, 8]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
        fontsize=9, loc="upper right")
plt.show()
```
